# LOADING DATA SETS

```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt

def load_data(file_path, filename):
    data = []
    csv_path = os.path.join(file_path, filename)
    with open(csv_path, 'r', newline='', encoding='utf-8') as csvfile:
        csv_reader = csv.DictReader(csvfile)
        for row in csv_reader:
            # Replace dash with 0 in the "Avg" column
            # changing in float
            # replace star(*) with space " "
            if 'Avg' in row:
                row['Avg'] = row['Avg'].replace('-', '0')
                row['Avg'] = float(row['Avg'])
            if 'HS' in row:
                row['HS'] = row['HS'].replace('*', '')    
            data.append(row)
    return data

print(os.getcwd())
```

# COMBINE AND FORMATING BATSMEN DATA

```python
def combineDataFramesWithSeasonCount(dataframes):
    player_stats = []
    
    for df in dataframes:
        for row in df:
            player_name = row['Player']
            existing_player_data = next((data for data in player_stats if data['Player'] == player_name), None)
            if existing_player_data is None:
                player_stats.append({
                    'Player': player_name,
                    'Runs': int(row.get('Runs', 0)),
                    'BF': int(row.get('BF', 0)),
                    'Mats': int(row.get('Mat', 0)),
                    'Inns': int(row.get('Inns', 0)),
                    'NO': int(row.get('NO', 0)),
                    '100': int(row.get('100', 0)),
                    '50': int(row.get('50', 0)),
                    '4s': int(row.get('4s', 0)),
                    '6s': int(row.get('6s', 0)),
                    'Seasons': 1,
                })
            else:
                existing_player_data['Runs'] += int(row.get('Runs', 0))
                existing_player_data['BF'] += int(row.get('BF', 0))
                existing_player_data['Mats'] += int(row.get('Mat', 0))
                existing_player_data['Inns'] += int(row.get('Inns', 0))
                existing_player_data['NO'] += int(row.get('NO', 0))
                existing_player_data['100'] += int(row.get('100', 0))
                existing_player_data['50'] += int(row.get('50', 0))
                existing_player_data['4s'] += int(row.get('4s', 0))
                existing_player_data['6s'] += int(row.get('6s', 0))
                existing_player_data['Seasons'] += 1

    for player_data in player_stats:
        # Calculate Strike Rate for each player
        if player_data['BF'] > 0:
            player_data['SR'] = round((player_data['Runs'] / player_data['BF']) * 100, 2)
        else:
            player_data['SR'] = 0
        
         # Calculate Batting Average for each player
        innings_played = player_data['Inns'] - player_data['NO']
        if innings_played > 0:
            player_data['Avg'] = round(player_data['Runs'] / innings_played, 2)
        else:
            player_data['Avg'] = 0

    return player_stats

def write_csv_data(output_folder, filename, data_list):
    if not data_list:
        print("No data to write.")
        return
    
    csv_path = os.path.join(output_folder, filename)

    with open(csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        # Get the fieldnames from the first element of the list
        fieldnames = data_list[0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for data in data_list:
            writer.writerow(data)

batting_list = [df_batting_2016, df_batting_2017, df_batting_2018, df_batting_2019,
                df_batting_2020, df_batting_2021, df_batting_2022]

player_batting_data = combineDataFramesWithSeasonCount(batting_list)

output_folder = "Outputs/Combined"
filename = "combined_batting_stats.csv"
write_csv_data(output_folder, filename, player_batting_data)
```

# COMBINE AND FORMATING BOWLERS DATA

```python
def combineDataWithSeasonCount(data_list):
    player_stats = {}

    for data in data_list:
        for row in data:
            player_name = row['Player']
            if player_name not in player_stats:
                player_stats[player_name] = {
                    'Player': player_name,
                    'Seasons': 1,
                    'Mat': int(row['Mat']),
                    'Inns': int(row['Inns']),
                    'Ov': float(row['Ov']),
                    'Runs': int(row['Runs']),
                    'Wkts': int(row['Wkts']),
                    'Avg': float(row['Avg']),
                    'Econ': float(row['Econ']),
                    'SR': float(row['SR']),
                    '4w': int(row['4w']),
                    '5w': int(row['5w'])
                }
            else:
                player_stats[player_name]['Seasons'] += 1
                player_stats[player_name]['Mat'] += int(row['Mat'])
                player_stats[player_name]['Inns'] += int(row['Inns'])
                player_stats[player_name]['Ov'] += float(row['Ov'])
                player_stats[player_name]['Runs'] += int(row['Runs'])
                player_stats[player_name]['Wkts'] += int(row['Wkts'])
                player_stats[player_name]['4w'] += int(row['4w'])
                player_stats[player_name]['5w'] += int(row['5w'])
                
                player_stats[player_name]['Avg'] = player_stats[player_name]['Runs'] / player_stats[player_name]['Wkts']
                player_stats[player_name]['Econ'] = player_stats[player_name]['Runs'] / player_stats[player_name]['Ov']
                player_stats[player_name]['SR'] = (player_stats[player_name]['Ov'] * 6) / player_stats[player_name]['Wkts']

    return list(player_stats.values())
def formatBowlingCombined(data_list):
    for row in data_list:
        row['Avg'] = round(row['Avg'] / row['Seasons'], 2)
        row['Econ'] = round(row['Econ'] / row['Seasons'], 2)
        row['SR'] = round(row['SR'] / row['Seasons'], 2)
    
    return data_list

def write_csv_data(output_folder, filename, data_list):
    csv_path = os.path.join(output_folder, filename)
    with open(csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = data_list[0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for row in data_list:
            writer.writerow(row)

bowling_list = [bowling_2016, bowling_2017,bowling_2018, bowling_2019,
                bowling_2020, bowling_2021, bowling_2022]
combined_data = combineDataWithSeasonCount(bowling_list)

player_bowling_data = formatBowlingCombined(combined_data)

# Write the combined and formatted bowling data to CSV
output_folder = "Outputs\Combined"
filename = "combined_and_formatted_bowling.csv"
write_csv_data(output_folder, filename, player_bowling_data)
```

# FINDING ALL_ROUNDERS DATA

```python
# Etracting all_rounders data from batting and bowling data sets.
def find_all_rounders(player_batting_data, player_bowling_data):
    all_rounders = []
    for batting_player in player_batting_data:
        player_name = batting_player['Player']
        for bowling_player in player_bowling_data:
            if bowling_player['Player'] == player_name:
                all_rounders.append({
                    'Player': player_name,
                    'Batting_Runs': batting_player.get('Runs', 0),
                    'Batting_Avg': batting_player.get('Avg',0),
                    'Batting_SR': batting_player.get('SR', 0),
                    'Batting_100': batting_player.get('100', 0),
                    'Batting_50': batting_player.get('50', 0),
                    'Batting_4s': batting_player.get('4s', 0),
                    'Batting_6s': batting_player.get('6s', 0),
                    'Bowling_Mat': bowling_player.get('Mat', 0),
                    'Bowling_Inns': bowling_player.get('Inns', 0),
                    'Bowling_Wkts': bowling_player.get('Wkts', 0),
                    'Bowling_Avg': bowling_player.get('Avg', 0),
                    'Bowling_Econ': bowling_player.get('Econ', 0),
                    'Bowling_SR': bowling_player.get('SR', 0),
                })
                break  # Found the player in both datasets, no need to continue searching

    return all_rounders

# Find all-rounders
player_all_rounders_data = find_all_rounders(player_batting_data, player_bowling_data)

# Create a CSV file for all-rounders
def write_all_rounders_data(output_folder, filename, player_all_rounders_data):
    csv_path = os.path.join(output_folder, filename)
    with open(csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = player_all_rounders_data[0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for data in player_all_rounders_data:
            writer.writerow(data)

output_folder = "Outputs/Combined"
filename = "all_rounders_data.csv"
write_all_rounders_data(output_folder, filename, player_all_rounders_data)
```

# SELECTING PLAYER FOR TEAM

```python
# Split the data into training and validation datasets
def split_data(data, train_fraction=0.8):
    np.random.shuffle(data)
    train_size = int(len(data) * train_fraction)
    train_data = data[:train_size]
    val_data = data[train_size:]
    return train_data, val_data

# Function to find all-rounders
def find_all_rounders(players_data):
    all_rounders = []
    for player in players_data:
        runs = player['Runs']
        strike_rate = player['SR']
        if 1500 <= runs <= 2500 and strike_rate > 145:
            all_rounders.append(player)
    return all_rounders

# Function to find batsmen
def find_batsmen(players_data):
    batsmen = []
    for player in players_data:
        runs = player['Runs']
        strike_rate = player['SR']
        if runs > 2500 and 125 <= strike_rate <= 200:
            batsmen.append(player)
    return batsmen

# Function to find bowlers
def find_bowlers(players_data):
    bowlers = []
    for player in players_data:
        wickets = player['Wkts']
        economy = player['Econ']
        average = player['Avg']
        if wickets >= 70 and economy < 5.0 and average < 4.0 :
            bowlers.append(player)
    return bowlers

# Select top players based on criteria
def select_top_players(players_data, criteria, num_players):
    sorted_players = sorted(players_data, key=lambda x: x[criteria], reverse=True)
    top_players = sorted_players[:num_players]
    return top_players

# Main function to form the cricket team
def form_cricket_team(player_batting_data, player_bowling_data):
    num_all_rounders = 2
    num_batsmen = 5
    num_bowlers = 4

    all_rounders = find_all_rounders(player_batting_data)[:num_all_rounders]
    batsmen = find_batsmen(player_batting_data)[:num_batsmen]
    bowlers = find_bowlers(player_bowling_data)[:num_bowlers]

    cricket_team = all_rounders + batsmen + bowlers
    return cricket_team

# Perform data splitting
train_batting_data, val_batting_data = split_data(player_batting_data)
train_bowling_data, val_bowling_data = split_data(player_bowling_data)

cricket_team = form_cricket_team(train_batting_data, train_bowling_data)

print("\nSelected  for the Team:")
print("-"*25)
for player in cricket_team:
    print(player['Player'])
print("-"*25)    
```

```python
# Load the player_batting_data from the CSV file
def load_csv_data(file_path):
    data = []
    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

file_path = r"C:\Users\sahga\Downloads\combined_and_formatted_batting.csv"
player_batting_data = load_csv_data(file_path)

# Prepare the data for regression
X = []
y = []
for player_data in player_batting_data:
    runs = int(player_data['Runs'])
    strike_rate = float(player_data['SR'])
    centuries = int(player_data['100'])
    fifties = int(player_data['50'])
    fours = int(player_data['4s'])
    sixes = int(player_data['6s'])

    X.append([1, strike_rate, centuries, fifties, fours, sixes])  # Adding 1 for bias term
    y.append(runs)

X = np.array(X)
y = np.array(y)

# Split the data into training and testing sets
def train_test_split(X, y, test_size=0.2, random_seed=None):
    if random_seed:
        np.random.seed(random_seed)
    num_samples = X.shape[0]
    num_test_samples = int(test_size * num_samples)
    test_indices = np.random.choice(num_samples, num_test_samples, replace=False)
    train_indices = np.array(list(set(range(num_samples)) - set(test_indices)))
    X_train, y_train = X[train_indices], y[train_indices]
    X_test, y_test = X[test_indices], y[test_indices]
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_seed=42)

# Create and train the regression model
def train_linear_regression(X, y):
    X_transpose_X_inv = np.linalg.inv(np.dot(X.T, X))
    X_transpose_y = np.dot(X.T, y)
    theta = np.dot(X_transpose_X_inv, X_transpose_y)
    return theta

theta = train_linear_regression(X_train, y_train)

# Make predictions on the test set
y_pred = np.dot(X_test, theta)

# Evaluate the model
mse = np.mean((y_test - y_pred)**2)
r2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - np.mean(y_test))**2))

print("Mean Squared Error:", mse)
print("R-squared:", r2)

# Use the model to predict runs for a new player
new_player_strike_rate = 120
new_player_centuries = 2
new_player_fifties = 5
new_player_fours = 100
new_player_sixes = 50

new_player_data = np.array([[1, new_player_strike_rate, new_player_centuries,
                             new_player_fifties, new_player_fours, new_player_sixes]])
predicted_runs = np.dot(new_player_data, theta)

print("Predicted Runs for the New Player:", predicted_runs[0])

```
